7) Finding divisors of a number using loops: 

A number $m$ is said to be a divisor of $n$ if $m$ divides $n$, or in other words the remainder of dividing $n$ by $m$ is $0$. For example, $5$ is a divisor of $10$.
In python, the modulus operator is used to find the remainder. The modulus operator is denoted by %. The expression n%m gives the remainder of division. For example:

    print(19%5)
    
will print $4$. 

**Example:** Execute the code below to find the remainder of dividing $271$ by $13$.

```python
n = 271
m = 13
print(n%m)
```

To check if a number is a divisor of another number we just check if the remainder is $0$.

We can use a loop to print all divisors of a given number. Either of while and for loops can be used. In the cell below, is a code to print all divisors of 30.

```python
n = 30
m = 1 
while m<=n:
    if n%m == 0:
        print(m)
    m = m+1
```
