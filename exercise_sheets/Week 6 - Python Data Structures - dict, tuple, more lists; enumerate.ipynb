For example, [2,0,1]  corresponds to the map which takes 0 to 2, 1 to 0 and 2 to 1. A permutation is called as a 1-cycle if starting with any one element and applying the permutation map repeatedly to the successive results one can get all the elements of the set. For example [2,0,1] is a 1-cycle as 0->2->1->0 so all elements are reached starting at 0. On the other hand [1,0,2] is not a 1-cycle as starting with 0, we cannot reach 2 by repeatedly applying the permutation. Find if the below permuations of 0, 1, 2, 3 are 1-cycles or not.

[1,2,3,0]
[2,1,0,3]
[3,2,1,0]
